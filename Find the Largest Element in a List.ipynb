Find the Largest Element in a List using for loop

```python
def find_largest_number(numbers_list: list[float]) -> float:
    """
    Args: 
        A list contain numeric values (int or float).
    Returns:
        Largest num in the list.
    Raises:
        TypeError: If input is not a list or contains non-numeric values.
        ValueError: If list is empty.
    """
    if not isinstance(numbers_list, list):
        raise TypeError("Input must be a list of numbers float or int")
    if not numbers_list:
        raise ValueError("List is not empty")
    for item in numbers_list:
        if not isinstance(item,(int,float)):
            raise TypeError("All values must be numeric")

    largest = numbers_list[0]
    for item in numbers_list:
        if item > largest:
            largest = item
    return largest
if __name__ == "__main__":
    try:
        user_input = input("Enter values in the list separate by space int or float only: ")
        numbers_list = [float(item) for item in user_input.split()]
        result = find_largest_number(numbers_list)
        print("Largest number in the list is: ", result)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
```
